## Primero importaré pandas y exportaré la base de datos

```python
import pandas as pd

df = pd.read_csv('https://breathecode.herokuapp.com/asset/internal-link?id=809&path=housing.csv')

df
```

## Debemos aplicar KMEANS sobre este conjunto de datos clasificandolo en 6 clusters

```python
from sklearn.cluster import KMeans

model = KMeans(n_clusters=6, random_state=42)

model.fit(df)

df['cluster'] = model.predict(df)

df
```

## AHora voy a ver las características promedio de cada grupo

```python
model.cluster_centers_
```

## Los centroides permiten identificar las zonas geográficas y niveles de ingreso característicos de cada cluster, facilitando la interpretación de los patrones encontrados en los datos. Por lo que vamos a graficar esta información y analizar los componentes principales (PCA)


```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

pca = PCA(n_components=3)

df_pca =pca.fit_transform(df.drop('cluster', axis=1))

df_pca = pd.DataFrame(df_pca, columns=['CP1', 'CP2', 'CP3'])

df_pca
```

## Ahora entrenaremos un modelo supervisado para predecir los clusters.

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

X = df.drop('cluster', axis=1)
y = df.cluster

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state=42)

model_s = KNeighborsClassifier(n_neighbors=6)
model_s.fit(X_train, y_train)
y_pred = model_s.predict(X_test)

accuracy_score(y_test, y_pred)
```

## Al entrenar con las etiquetas obtenidas en KMEANS y luego con KNEIGHBORS, nuestra presición será del 99.90%, los cuales pueden trabajar conjuntamente porque KMEANS termina siendo la versión no supervisada de KNEIGHBORS, y tienen la misma lógica de puntos por distancia.
